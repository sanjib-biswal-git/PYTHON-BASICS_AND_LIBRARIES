Seaborn is a Python visualization library built on top of Matplotlib.

Its main advantage: it creates beautiful, statistical plots with very little code.

Works natively with Pandas DataFrames (so no need to manually extract columns).    

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")  # Sample dataset (restaurant bills)
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.show()
```

# Working with Titanic data set

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset("titanic")
print(titanic.head())
print(titanic.info())
```

```python
#Distribution Plots

# Histogram
sns.histplot(titanic["age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# KDE plot
sns.kdeplot(data=titanic, x="age", hue="sex", fill=True)
plt.title("Age Density by Gender")
plt.show()
```

```python
#Categorical Plots

# Countplot: Number of passengers by class
sns.countplot(data=titanic, x="class", hue="sex")
plt.title("Passenger Count by Class & Gender")
plt.show()

# Barplot: Average fare by class
sns.barplot(data=titanic, x="class", y="fare", hue="sex")
plt.title("Average Fare by Class & Gender")
plt.show()

# Boxplot: Age distribution across classes
sns.boxplot(data=titanic, x="class", y="age", hue="sex")
plt.title("Age Distribution by Class & Gender")
plt.show()

# Violinplot: Age distribution across classes
sns.violinplot(data=titanic, x="class", y="age", hue="sex", split=True)
plt.title("Violin Plot of Age by Class & Gender")
plt.show()
```

```python
#Relational Plots

# Scatterplot: Age vs Fare
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex")
plt.title("Scatter: Age vs Fare")
plt.show()

# Lineplot: Survival by Age
sns.lineplot(data=titanic, x="age", y="survived", hue="sex", errorbar=None)
plt.title("Survival vs Age (by Gender)")
plt.show()
```

```python
#Heatmaps

corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot
sns.pairplot(titanic[["age","fare","survived"]].dropna(), hue="survived")
plt.show()

# Jointplot: Age vs Fare
sns.jointplot(data=titanic, x="age", y="fare", kind="hex", cmap="magma")
plt.show()
```
