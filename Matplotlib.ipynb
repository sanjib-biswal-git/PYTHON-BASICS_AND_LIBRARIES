```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()

```
