# Implementing for a DataSet


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample dataset
x = np.arange(1, 11)
y = x ** 2
z = np.random.randint(1, 20, size=10)
```

```python
#Basic Plots

# Line plot
plt.plot(x, y)
plt.title("Line Plot: y = x^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Scatter plot
plt.scatter(x, z, color='red', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Random values")
plt.show()

# Bar plot
plt.bar(x, z, color='green')
plt.title("Bar Plot")
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.show()

# Histogram
plt.hist(z, bins=5, color='orange')
plt.title("Histogram")
plt.show()

# Pie chart
sizes = [30, 20, 25, 25]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
#Customization


plt.plot(x, y, color="purple", linestyle="--", marker="o", label="y=x^2")
plt.plot(x, x*3, color="blue", linestyle="-.", marker="s", label="y=3x")
plt.title("Customized Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#SubPlots

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0,0].plot(x, y)
axs[0,0].set_title("Line")

axs[0,1].scatter(x, z, color='red')
axs[0,1].set_title("Scatter")

axs[1,0].bar(x, z)
axs[1,0].set_title("Bar")

axs[1,1].hist(z)
axs[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```
